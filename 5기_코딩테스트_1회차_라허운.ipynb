## 1. 문자열 나누기

image.png

```python
import time 

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    answer = 0
    list = []
    for i in range(len(s)):
        if len(list) == 0:
            answer +=1
            list.append(s[i])
            #print(list)
            continue
        if list[-1] == s[i]:
            list.append(s[i])
        else:
            list.pop()


# ==============================
    return answer

print(solution("banana"))
print(solution("abracadabra"))
print(solution("aaabbaccccabba"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")

```

# 4.두 개 뽑아서 더하기

image.png
image.png

```python
import time 

start = time.time()

def solution(s):
# 답안 작성 부분 ===============
    answer = []

    for i in range(len(s)):
        for j in range(i+1, len(s)):
            answer.append(s[i] + s[j])
    answer = sorted(set(answer))

# ==============================
    return answer

print(solution([2,1,3,4,1]))
print(solution([5,0,2,7]))

end = time.time()


print("작동시간:", f"{end - start:.5f} sec")
# return [2,5,7,9,12]
```

#5.	최대공약수와 최소공배수

image.png
image.png

```python
import time 
from fractions import gcd

start = time.time()

def solution(n,m):
# 답안 작성 부분 ===============
    g = gcd(n,m)
    lcm = (n*m)//g
    

# ==============================
    return [g,lcm]

print(solution(3,12))
print(solution(2,5))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
