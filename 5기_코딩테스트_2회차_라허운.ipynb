# 3. 핸드폰 번호 가리기

image.png

```python
import time 

start = time.time()

def solution(phone_number):
  answer = ''
# 답안 작성 부분 == ==============
  for i in range(len(phone_number)):
    if i < len(phone_number) - 4:
      answer += '*'
    else:
      answer += phone_number[i]

# ==============================
  return answer

print(solution("01033334444"))
print(solution("027778888"))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
