# 2. 소수 찾기

image.png

```python
import time 

start = time.time()

def solution(n):
# 답안 작성 부분 ===============
    answer =0
    for i in range(2, n+1):
        if i == 2:
            answer += 1
            continue

        for j in range(2, int(i ** (1/2)) + 1):
            if i % j == 0:
                answer -= 1
                continue
        answer += 1

        



# ==============================
    return answer

print(solution(10))
print(solution(5))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

```python
import time 

start = time.time()

def solution(number, limit, power):
# 답안 작성 부분 ===============
    answer = 0
    for num in range(1,number+1):
        cnt = 0
        for div in range(1,int(num**0.5)+1):
            if num%div == 0:
                cnt+=2
        if (num**0.5)%1 == 0:
            cnt-=1
        if cnt<=limit:
            answer+=cnt
        else:
            answer+=power

# ==============================
    return answer


print(solution(5, 3, 2))
print(solution(10, 3, 2))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

activation function은 시그모이드이며 최종 아웃풋의 출력값을 구하시오

```python
import numpy as np

def sigmoid(x):
    return 1/(1+np.exp(-x))


def identity_function(x):
    return x

X = np.array([1.0, 0.5])
W1 = np.array([[0.1, 0.3, 0.5],[0.2, 0.4, 0.6]])
B1 = np.array([0.1, 0.2, 0.3])

A1 = np.dot(X, W1) + B1


Z1 =(sigmoid(A1))  # 함수와 괄호안에 답 넣기

W2 = np.array([[0.1, 0.4], [0.2, 0.5], [0.3, 0.6]])
B2 = np.array([0.1, 0.2])

A2 = np.dot(Z1, W2) + B2     # 수식
Z2 =(sigmoid(A2))  # 함수와 괄호안에 답 넣기



W3 = np.array([[0.1, 0.3], [0.2, 0.4]])
B3 = np.array([0.1, 0.2])

A3 = np.dot(Z2, W3) + B3                   # 수식
Y = identity_function(A3) 
print(Y)     # array([0.31682708, 0.69627909]) 나와야함.
```
